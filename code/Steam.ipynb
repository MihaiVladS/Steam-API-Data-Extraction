```python
def get_request(url, parameters=None):
    counter = 0
    while counter < 10:
        response = requests.get(url=url, params=parameters)
        if response:
            return response.json()
        else:
            time.sleep(5)
            return get_request(url, parameters)
    time.sleep(5)
    counter += 1
```
